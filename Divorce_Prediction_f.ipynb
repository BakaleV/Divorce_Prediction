```python
data = raw_data.reset_index(drop = True)
data.head()
```

```python
targets = data['Divorce_Y_N']
inputs = data.drop(['Divorce_Y_N'],axis = 1)
inputs.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs,targets,test_size = 0.2,random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier


KNC = KNeighborsClassifier(n_neighbors=3)
KNC.fit(x_train, y_train)
```

```python
y_predictor = KNC.predict(x_test)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predictor)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predictor))
```

```python
#split the dataset into train, validation, and test
sample_count = inputs.shape[0]
train_count = int(0.8*sample_count)
validation_count = int(0.1*sample_count)
test_count = int(sample_count - validation_count - train_count)
print(sample_count,train_count,validation_count,test_count)
```

```python
train_inputs = inputs[:train_count]
validation_inputs = inputs[train_count:validation_count + train_count]
test_inputs = inputs[validation_count + train_count :]

train_targets = targets[:train_count]
validation_targets = targets[train_count:validation_count + train_count]
test_targets = targets[validation_count + train_count :]
```

```python
#save the data in .npz files
np.savez('Divorce_data_train',inputs = train_inputs,targets = train_targets)
np.savez('Divorce_data_validation',inputs = validation_inputs,targets = validation_targets)
np.savez('Divorce_data_test',inputs = test_inputs,targets = test_targets)
```
